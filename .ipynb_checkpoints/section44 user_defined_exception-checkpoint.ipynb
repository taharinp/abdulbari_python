```python
def div(a,b):
    if b!=0:
        

      c=a//b
      return c
    else:  
     raise ZeroDivisionError
        


try:
    res=div(10,0)
    print(res)
except:


  print('division by zero')
```

```python
def fun():
    try:
        x=int('24')
        return x
    except Exception as e:
      raise e
    finally:

      print('end')


res =fun()
print(res)
```
