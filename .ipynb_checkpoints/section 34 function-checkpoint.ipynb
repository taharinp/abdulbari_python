max 3

```python
def max3(a, b, c,/):
    if a >b and a > c:
        return a
    elif b > c:
        return b
    else:
        return c

print(max3(10,12,5))

```

pangram phase

```python

import re

def pangram(phrase):
    letters = re.sub(r'[^a-zA-Z]', '', phrase)
    letter_set = set(letters.lower())
    if len(letter_set) == 26:
        return True
    else:
        return False

str = 'The quick brown fox jumps over the lazy dog'

print(pangram(str))
```
