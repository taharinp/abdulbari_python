```python
def fun(*args):
    for x in args:
        if type(x)==int:
           print(x)
fun(12,3.4,'jgj')
```

```python
def fun(a,b,*args):
    print(a,b,args)
fun(10,23,34,67,8)    
```

```python
def fun(a,b,c):
    sum=a+b+c
    prod=a*b*c
    return sum,prod
print(fun(2,3,4))    
```

```python
def result(m1,m2,m3):
    total=m1+m2+m3
    avg=total/3
    if avg>=45:
        grade='pass'
    else:
        grade='fail'
    return total,avg,grade 
print(result(23,43,56))
    
```
