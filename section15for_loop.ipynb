3. fibonacci series


```python

n=int(input('enter number'))

a=0
b=1
for i in range(0,n):
    c=a+b
    a=b
    b=c
    print(a)
```

5.prime number

```python
n=int(input('enter number'))

count=0

for i in range(1,n+1):
    if n%i==0:
        count +=1
if count==2:
    print('prime')
else:
    print('no prime')
    
```
