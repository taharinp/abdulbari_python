```python
import math

class Circle:
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2

    def perimeter(self):
        return 2 * math.pi * self.radius

# Hardcoded radius value
radius = 5.0

c1 = Circle(radius)
print(f"Area: {c1.area():.2f}")
print(f"Perimeter: {c1.perimeter():.2f}")
```

```python
class BankAccount:

    # Constructor
    def __init__(self, name, balance):
        self.name = name
        self.balance = balance

    # Deposit method
    def deposit(self, amount):
        self.balance = self.balance + amount
        print("Deposited:", amount)

    # Withdraw method
    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance = self.balance - amount
            print("Withdrawn:", amount)
        else:
            print("Insufficient Balance")

    # Display balance
    def show_balance(self):
        print("Name:", self.name)
        print("Balance:", self.balance)


# Creating object
acc1 = BankAccount("Rahul", 1000)

acc1.show_balance()

acc1.deposit(500)
acc1.withdraw(300)

acc1.show_balance()
```

person profile using property methods

```python
class Person:

    # Constructor
    def __init__(self, name, age):
        self._name = name
        self._age = age

    # Getter method
    @property
    def age(self):
        return self._age

    # Setter method
    @age.setter
    def age(self, value):
        if value > 0:
            self._age = value
        else:
            print("Invalid Age")

    # Getter for name
    @property
    def name(self):
        return self._name

    # Display method
    def show(self):
        print("Name:", self._name)
        print("Age:", self._age)


# Object creation
p1 = Person("Rahul", 20)

p1.show()

# Using property
p1.age = 25

print("Updated Age:", p1.age)
```

```python
class Temperature:

    # Celsius to Fahrenheit
    @staticmethod
    def c_to_f(c):
        f = (c * 9/5) + 32
        return f

    # Fahrenheit to Celsius
    @staticmethod
    def f_to_c(f):
        c = (f - 32) * 5/9
        return c


# Using static methods
print("Celsius to Fahrenheit:", Temperature.c_to_f(30))

print("Fahrenheit to Celsius:", Temperature.f_to_c(86))
```
