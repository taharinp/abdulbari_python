count no of digits

```python
n=int(input('enter number'))
i=0
while n>0:
    i=i+1
    n=n//10
    print('number of digits',i)
```

```python
n=int(input('enter a number'))
rev=0
while n>0:
    r=n%10
    rev=rev*10+r
    n=n//10
    print('reverse number:',rev)
```

PALINDROME

```python
n=int(input('enter number'))
rev=0
m=n
while n>0:
    r=r%10
    rev=rev*10+r
    n=n//10
if rev==m:
    print('palindrome')
else:
    print('not palindrome')
```
